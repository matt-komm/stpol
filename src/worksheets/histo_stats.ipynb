```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
events = []
seconds = []
fi = open("stats")
for line in fi.readlines():
    spl = line.split()
    ev = int(spl[1])
    sec = int(spl[5])
    events.append(ev)
    seconds.append(sec)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
ax.scatter(events, seconds, marker='.')
ax.set_xlim(0, 0.6*10**7)
ax.set_xlabel("number of events")
ax.set_ylabel("processing time (sec)")
```

```python
fig = plt.figure()
ax = plt.axes()
ax.hist(seconds, bins=40, color='red', histtype='step')
ax.set_xlabel("processing time (seconds)")
ax.semilogy()
```

```python
weights = zip(["puw__nom", "puw__up", "puw__down"], Weights.pu_syst)
weights.append(("unw", "1.0"))
baseline = Cuts.mt_mu()*Cuts.lepton("mu")*Cuts.hlt("mu")*Cuts.rms_lj
cuts = [
    ("2j0t", baseline*Cuts.n_jets(2)*Cuts.n_tags(0)),
    ("2j1t", baseline*Cuts.n_jets(2)*Cuts.n_tags(1)*Cuts.top_mass_sig*Cuts.eta_lj),
    ("3j1t", baseline*Cuts.n_jets(3)*Cuts.n_tags(1)),
    ("3j2t", baseline*Cuts.n_jets(3)*Cuts.n_tags(2)),
]
```

```python
htot = {}
htot_cos = {}
skips = {
    "2j0t":["ttjets", "tchan"],
    "2j1t":["ttjets", "wjets"],
    "3j1t":["wjets", "tchan"],
    "3j2t":["wjets", "tchan"]
}

for cutn, cut in cuts:
    for k, smps in samps.items():
        if k in skips[cutn]:
            continue
        for wn, w in weights if k!="data" else [("unw", "1.0")]:
            hists = []
            hists_cos = []
            for x in smps:
                hist = x.drawHistogram(
                    "n_vertices", str(cut),
                    weight=str(w), binning=[40, 0, 40]
                )
                hist_cos = x.drawHistogram(
                    "cos_theta", str(cut),
                    weight=str(w), binning=[40, -1, 1]
                )
                if k!="data":
                    hist.Scale(x.lumiScaleFactor(lumi))
                    hist_cos.Scale(x.lumiScaleFactor(lumi))
                hists.append(hist)
                hists_cos.append(hist_cos)
            htot["/".join([cutn, wn, k])] = reduce(lambda x,y: x+y, hists)
            htot_cos["/".join([cutn, wn, k])] = reduce(lambda x,y: x+y, hists_cos)
```

```python
def ratio_subplots():
    gs = GridSpec(2, 1,
        width_ratios=[1],
        height_ratios=[4,1]
    )
    ax1 = plt.subplot(gs[0])
    ax2 = plt.subplot(gs[1])
    return ax1, ax2

def axes_style(ax):
    ax.grid(True, which='both')
    ax.tick_params(axis='both', which='major', labelsize=16)

def errs_y(nominal, up, down):
    bins_nom = numpy.array(list(nominal.y()))
    bins_up = numpy.array(list(up.y()))
    bins_down = numpy.array(list(down.y()))
    return numpy.abs(bins_nom-bins_up), numpy.abs(bins_nom-bins_down)

def ratio(data, mc):
    return (mc - data)/data
```

```python
sampnames = {
    "ttbar":r"$\mathbf{t \bar{t}}$",
    "tchan":r"$\textrm{signal (t-channel)}$",
    "wjets":r"$W (\rightarrow l \nu) + \textrm{jets}$",

}

cutnames = {
    "2j1t": r"2J1T, $M_{\textrm{top}} \in [130,220], |\eta_{j'}|>2.5$",
    "2j0t": "2J0T",
    "3j1t": "3J1T",
    "3j2t": "3J2T",
}
def draw(cut, proc):
    fig = plt.figure(dpi=500, figsize=(10, 10))
    ax1, ax2 = ratio_subplots()
    map(axes_style, [ax1, ax2])
    
    ax1.set_title(r"PU weighting for %s in %s" % (sampnames[proc], cutnames[cut]), fontsize=16)
    ax1.tick_params(axis='both', which='major', labelsize=16)
    ax1.set_ylabel("normalized \# of events", fontsize=16)
    
    def h(t, proc=proc):
        return htot["%s/%s/%s" % (cut, t, proc)]
    hist_err(ax1, h("unw", "data"), color="black", fmt='o', label="data")
    hist_err(ax1, h("unw"), fmt='--', label="%s unweighted" % sampnames[proc])
    hist_err(ax1, h("puw__nom"),
        yerr=errs_y(
            h("puw__nom"),
            h("puw__up"),
            h("puw__down")
        ),
    fmt='-', linewidth=2, label="%s PU weighted" % sampnames[proc])
    
    leg = ax1.legend(fontsize=16)
    
    hratios = {}
    
    hratios["unw"] = ratio(h("unw", "data"), h("unw"))
    hratios["puw"] = ratio(h("unw", "data"), h("puw__nom"))
    
    ax2.set_ylim([-1.0, 1.0])
    ax2.grid(True, which='both')
    ax2.axhline(0, color='black')
    ax2.set_xlabel(r"$N_{\textbf{vertices}}$", fontsize=16)
    ax2.tick_params(axis='both', which='major', labelsize=16)
    ax2.set_ylabel("(exp. - meas.)/meas.", fontsize=16)
    
    # legitems = []
    hist_err(ax2, hratios["unw"], fmt='--')
    
    hist_err(ax2, hratios["puw"],
        yerr=
            errs_y(hratios["puw"], 
                ratio(h("unw", "data"), h("puw__up")),
                ratio(h("unw", "data"), h("puw__down"))
            )
        , fmt='-', linewidth=2
    )
    try:
        import os
        os.makedirs("out/plots/puw")
    except:
        pass
    plt.savefig("out/plots/puw/%s.pdf" % cut)

#leg3 = legend(["unweighted", "PU weighted"])
```

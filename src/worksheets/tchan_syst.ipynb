```python
channel = "mu"
bn = "/hdfs/local/stpol/step3/37acf5_343e0a9_Aug22/" + channel
inf = [
    bn + "/mc/iso/nominal/Jul15/T_t_ToLeptons.root",
    bn + "/mc/iso/nominal/Jul15/Tbar_t_ToLeptons.root",
    bn + "/mc_syst/iso/SYST/Jul15/Tbar_t_ToLeptons_scaledown.root",
    bn + "/mc_syst/iso/SYST/Jul15/T_t_ToLeptons_scaledown.root",
    bn + "/mc_syst/iso/SYST/Jul15/Tbar_t_ToLeptons_scaleup.root",
    bn + "/mc_syst/iso/SYST/Jul15/T_t_ToLeptons_scaleup.root",
]
samps = map(Sample.fromFile, inf)
```

```python
hists = {}
cut = Cuts.lepton(channel)*Cuts.metmt(channel)*Cuts.rms_lj*Cuts.hlt(channel)*Cuts.n_jets(2)*Cuts.n_tags(1)*Cuts.mva_wp(channel)
for s in samps:
    hists[s.name] = s.drawHistogram("cos_theta", str(cut), binning=[60, -1, 1])
    hists[s.name].Scale(s.lumiScaleFactor(20000))
    print s.name, hists[s.name].GetEntries()
```

```python
pretty={
    "Tbar_t_ToLeptons": r"t-channel (antitop)",
    "T_t_ToLeptons": r"t-channel (top)",
}
def drawshape(base):
    hc = deepcopy(hists)
    for h in hc.values():
        h.Rebin(4)
    scaleup = hc[base + "_scaleup"]/hc[base]
    scaledown = hc[base + "_scaledown"]/hc[base]
    nominal = hc[base]/hc[base]
    z = numpy.zeros(len(scaleup))
    fig = plt.figure(figsize=(10,10))
    ax = plt.axes()
    axes_style(ax)
    if scaleup.GetEntries()<50:
        for i in range(1,scaleup.nbins()+1):
            scaleup.SetBinContent(i, 0)
    hist_err(ax, scaleup, yerr=(z,z), label="scale up", linewidth=2)
    hist_err(ax, nominal, yerr=(z,z), label="nominal", linewidth=2)
    hist_err(ax, scaledown, yerr=(z,z), label="scale down", linewidth=2)
    ax.set_title(r"$Q^2$ variations for t-channel, %s, 2J1T, MVA, %s" % (pretty[base], channel))
    ax.set_ylabel("variation / nominal", fontsize=16)
    ax.set_xlabel(r"$\cos \theta^*$", fontsize=24)
    ax.annotate("Statistics: \r\n nominal: %d \r\n scale up: %d \r\n scale down: %d" % (hc[base].GetEntries(), hc[base+"_scaleup"].GetEntries(), hc[base+"_scaledown"].GetEntries()),
        xy=(0.47, 0.87), xycoords="axes fraction", fontsize=16
    )
    ax.set_ylim([0.5, 1.5])
    ax.legend(fontsize=16)
    return ax
```

```python
drawshape("T_t_ToLeptons")
```

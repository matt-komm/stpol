```python
import ROOT
from plots.common.mpl_hists import draw_hists,hist_err, axes_style
from plots.common.mpl_tdrstyle import plt
from rootpy.io import File
%matplotlib inline
```

```python

def loadh(fn, title):
    
    f1 = File(fn)
    ROOT.gROOT.cd()
    hi1 = f1.Get("cos_theta_weighted").Clone()
    hi1.Rebin(5)
    hi1.SetTitle(title)
    hi1.Scale(1.0/hi1.Integral())
    return hi1
    
hi1 = loadh("results/gen_level_cos_theta/hmass.root", "Sherpa massive")
hi2 = loadh("results/gen_level_cos_theta/hnomass.root", "Sherpa massless")
hi3 = loadh("results/gen_level_cos_theta/hmadgraph.root", "MadGraph massive")
chi2 = hi1.Chi2Test(hi2, "WW CHI2/NDF")
pks = hi1.KolmogorovTest(hi2)
```

```python
def sciform(v):
    return ("%.2E" % v).replace("E", r" \times 10^{") + "}"
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes()
axes_style(ax);
an1 = ax.annotate(
    "Sherpa massive / massless \n compatibility:\n" + 
    r"\chi^2/\nu = %.2f" % chi2 + "\n" + r"p_{KS} = %s" % sciform(pks),
    xy=(0.05, 0.05), xycoords='axes fraction', fontsize=16, backgroundcolor='w'
);
s = ("Entries: \n %s = %d \n" % (hi1.GetTitle(), hi1.GetEntries()) +
    "%s = %d \n" % (hi2.GetTitle(), hi2.GetEntries()) +
    "%s = %d" % (hi3.GetTitle(), hi3.GetEntries()))
#s = "ASd"
an2 = ax.annotate(
    s, 
    xy=(0.60, 0.05), xycoords='axes fraction', fontsize=16, backgroundcolor='w'
);
hist_err(ax, hi1, label=hi1.GetTitle(), linewidth=2);
hist_err(ax, hi2, label=hi2.GetTitle(), linewidth=2);
hist_err(ax, hi3, label=hi3.GetTitle(), linewidth=2);
ax.set_ylim([0.0, 0.13]);
ax.set_title(r"gen-level comparison for W+jets massive vs. massless b-quark", fontsize=16);
ax.set_xlabel(r"$\cos \theta^*$", fontsize=24);
#ax.set_xlabel(r"$p_{t, l}$ \textrm{(transverse momentum of the light quark)}", fontsize=16)
#ax.set_xlabel(r"$p_{t, b}$ \textrm{(transverse momentum of the b quark)}", fontsize=16)
ax.set_ylabel("normalized events", fontsize=16);
leg = legend(loc=1, fontsize=16);
#ax.semilogy()
```

```python
def sumh(h):
    return reduce(lambda x,y:x+y, h)
def draw(_hists, var):
    hists = deepcopy(_hists)
    unw = sumh(hists[var]["1.0"].values())
    w =  sumh(hists[var]["wjets_mg_flavour_shape_weight"].values())
    for h in [unw, w]:
        h.Rebin(3)
    fig = plt.figure(figsize=(10,10))
    ax = plt.axes()
    ax.set_title("BDT input variable dependence of W+jets reweighting, 2J1T, %s channel" % channel)
    axes_style(ax)
    ax.set_xlabel(varnames[var], fontsize=16)
    ax.set_ylabel("Events / $20 fb^{-1}$", fontsize=16)
    z = numpy.zeros(len(unw))
    hist_err(ax, unw, label="unweighted", linewidth=2)
    hist_err(ax, w, yerr=(z,z), label="weighted", linewidth=2)
    ax.legend(fontsize=16)
    return ax
```
